## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Line Plot

<p><strong>Explanation:</strong> A line plot is used to display data points connected by a line. Itâ€™s useful for showing trends over time or continuous data.</p>
<p><strong>Usage:</strong> Typically used when you have continuous data, such as changes in a variable over time (e.g., stock prices, temperatures).</p>

```python
# Create some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Modified Line plot with markers and grid style
plt.figure(figsize=(10,6))  # Slightly larger figure size
plt.plot(x, y, linestyle='-.', color='blue', marker='o', markersize=5, label='sin(x)')  # Added markers
plt.title('Enhanced Sine Wave Plot', fontsize=16)  # Modified title with larger font
plt.xlabel('X Axis (Radians)', fontsize=12)  # Modified X label with description
plt.ylabel('Y Axis (Amplitude)', fontsize=12)  # Modified Y label with description
plt.legend(loc='upper right')  # Moved legend position to upper right
plt.grid(True, which='both', linestyle='--', linewidth=0.7)  # Enhanced grid with dashed lines and grid for both axes
plt.tight_layout()  # Adjusts layout for better display
plt.show()
```

## 2. Bar Plot

<p><strong>Explanation:</strong> A bar plot is used to display data in rectangular bars where the height of each bar is proportional to the value it represents.</p>
<p><strong>Usage:</strong> Ideal for comparing quantities across different categories, like sales figures for different products, or populations in different regions.</p>

```python
# Data for Bar plot
categories = ['Category A', 'Category B', 'Category C']
values = [15, 35, 25]

# Enhanced Bar plot with additional customization
plt.figure(figsize=(10,6))  # Increased figure size
plt.bar(categories, values, color='DarkBlue', hatch='', edgecolor='black', linewidth=1.5)  # Changed color and hatch pattern
plt.title('Enhanced Bar Plot', fontsize=16)  # Enlarged title font size
plt.xlabel('Categories', fontsize=12)  # Increased font size of X label
plt.ylabel('Values', fontsize=12)  # Increased font size of Y label
plt.ylim(0, 40)  # Added Y-axis limit for better scaling
plt.grid(True, axis='y', linestyle='--', alpha=0.7)  # Added grid lines on Y-axis
plt.tight_layout()  # Ensures the layout is well adjusted
plt.show()
```

## 3. Histogram

<p><strong>Explanation:</strong> A histogram is a graphical representation of the distribution of numerical data. It groups data into bins (intervals) and shows how many data points fall into each bin.</p>
<p><strong>Usage:</strong> Useful for showing the distribution of data (e.g., exam scores, heights of people) and detecting patterns like skewness or outliers.</p>

```python
# Data for histogram
data = np.random.randn(1000)

# Enhanced Histogram with additional customizations
plt.figure(figsize=(10,6))  # Increased figure size
plt.hist(data, bins=30, color='lightblue', edgecolor='black', alpha=0.7, histtype='bar', rwidth=0.9)  # Changed color, added transparency and width adjustment
plt.title('Enhanced Histogram', fontsize=16)  # Enlarged title font size
plt.xlabel('Value', fontsize=12)  # Increased X-axis label font size
plt.ylabel('Frequency', fontsize=12)  # Increased Y-axis label font size
plt.grid(True, linestyle='--', alpha=0.6)  # Added grid lines with slight transparency
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```

## 4. Scatter Plot

<p><strong>Explanation:</strong> A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axes indicates values for an individual data point.</p>
<p><strong>Usage:</strong> Commonly used to determine the relationship or correlation between two variables (e.g., height vs weight).</p>

```python
# Data for scatter plot
x = np.random.rand(100)
y = np.random.rand(100)

# Enhanced Scatter plot with additional customization
plt.figure(figsize=(10,6))  # Increased figure size
plt.scatter(x, y, color='purple', s=120, edgecolor='black', alpha=0.75, marker='o')  # Changed color, marker size, and added transparency
plt.title('Enhanced Scatter Plot', fontsize=16)  # Increased title font size
plt.xlabel('X Axis', fontsize=12)  # Enlarged X-axis label font
plt.ylabel('Y Axis', fontsize=12)  # Enlarged Y-axis label font
plt.grid(True, linestyle='--', alpha=0.6)  # Added grid lines for better visualization
plt.tight_layout()  # Ensure layout is adjusted properly
plt.show()
```

## 5. Heatmap (Using Seaborn)

<p><strong>Explanation:</strong> A heatmap is a two-dimensional visualization where individual values in a matrix are represented as colors.</p>
<p><strong>Usage:</strong> Useful for visualizing data density, correlations, or patterns in a dataset (e.g., showing correlations between variables in a correlation matrix).</p>

```python
# Data for heatmap
data = np.random.rand(10, 12)

# Enhanced Heatmap with additional customizations
plt.figure(figsize=(12,8))  # Increased figure size
sns.heatmap(data, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, linecolor='black', cbar_kws={'shrink': 0.8})  # Changed colormap, added lines, and customized color bar
plt.title('Enhanced Heatmap', fontsize=16)  # Enlarged title font size
plt.tight_layout()  # Ensure proper layout
plt.show()
```

## 6. Pair Plot (Using Seaborn with a sample dataset)

<p><strong>Explanation:</strong> A pair plot allows you to plot pairwise relationships between different numerical variables in a dataset. It creates a matrix of scatter plots for each combination of variables.</p>
<p><strong>Usage:</strong> Helpful for exploring relationships between multiple variables at once, especially in large datasets (e.g., analyzing correlations in the Iris dataset).</p>

```python
# Load a dataset
iris = sns.load_dataset('iris')

# Enhanced Pair Plot with additional customizations
plt.figure(figsize=(12,8))  # Adjust figure size
sns.pairplot(iris, hue='species', palette='viridis', markers=["o", "s", "D"], diag_kind='kde', height=2.5)  # Changed color palette, markers, and added KDE on diagonal
plt.suptitle('Enhanced Iris Pair Plot', y=1.02, fontsize=16)  # Added title with adjusted position and size
plt.tight_layout()  # Ensure the layout is properly adjusted
plt.show()
```

### The diagonal elements of the grid show the distribution (density plot) of each feature for the three species.
Example: The top-left plot shows the distribution of sepal length for Setosa, Versicolor, and Virginica.
### The off-diagonal plots represent the scatter plots for each pair of features.
Example: The plot in the second row and first column shows how sepal width relates to sepal length.
### Each point in the scatter plots represents an individual flower sample, with the color indicating the species.

## 7. Pie Chart

<p><strong>Explanation:</strong> A pie chart represents data in a circular graph, where each slice corresponds to a category's contribution to the total.</p>
<p><strong>Usage:</strong> Useful for showing proportional data, like market share, or parts of a whole in percentage.</p>
<pre>

```python
# Pie Chart Data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [35, 15, 20, 30]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Explode 1st slice

# Enhanced Pie Chart with additional customizations
plt.figure(figsize=(8,8))  # Slightly increased figure size
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140,
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})  # Adjusted start angle, edge colors, and formatting

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Enhanced Pie Chart', fontsize=16)  # Added title with a larger font size
plt.tight_layout()  # Adjust layout to ensure everything fits well
plt.show()
```

## 8. Violin Plot

<p><strong>Explanation:</strong> A violin plot combines the features of a box plot and a density plot, showing both the distribution of the data and the probability density. The thicker parts of the violin represent areas where the data points are more concentrated.</p>
<p><strong>Usage:</strong> Useful for comparing the distribution of data across different categories, especially when dealing with multimodal distributions (data with more than one peak).</p>

```python
# Load a dataset
iris = sns.load_dataset('iris')

# Enhanced Violin Plot with additional customizations
plt.figure(figsize=(10,6))  # Slightly increased figure size
sns.violinplot(x="species", y="sepal_length", data=iris, palette="coolwarm", inner="quartile", linewidth=1.3)  # Changed palette, added quartiles, increased line width
plt.title("Enhanced Violin Plot: Sepal Length by Species", fontsize=16)  # Added more descriptive title with larger font
plt.xlabel("Species", fontsize=12)  # Increased X-axis label font size
plt.ylabel("Sepal Length", fontsize=12)  # Increased Y-axis label font size
plt.grid(True, linestyle='--', alpha=0.6)  # Added grid lines for better readability
plt.tight_layout()  # Ensure the layout is well adjusted
plt.show()
```

### 1. Violin Shape:
### The outer shape of each violin represents the density estimate of the data. This shows where the data points are more concentrated or sparse.
Wider parts of the violin indicate areas where data points are more densely packed (higher frequency), while narrower parts indicate lower density (fewer data points).
Box Plot Elements Inside the Violin:

### 2. Black bar (box): This indicates the interquartile range (IQR), which is the range between the 25th and 75th percentiles.
Black lines (whiskers): These extend from the box to show the range of the data, except for outliers.
### 3. Species Comparison:
Each violin represents the distribution of sepal length for one species of iris (Setosa, Versicolor, and Virginica).

### 4. Density and Symmetry:
The shape of each violin helps show whether the data is symmetric or skewed.
For example, Setosa shows a more symmetric distribution, while Versicolor has a slight skew, with more data concentrated towards lower values.

## 9. Swarm Plot

<p><strong>Explanation:</strong> A swarm plot is a categorical scatterplot where the points are adjusted to avoid overlapping, allowing for the visualization of individual data points along with their distribution.</p>
<p><strong>Usage:</strong> Ideal for showing individual data points and how they are distributed across categories. Itâ€™s especially useful when you want to visualize clusters or patterns within the data.</p>

```python
# Load a dataset
iris = sns.load_dataset('iris')

# Enhanced Swarm Plot with additional customizations
plt.figure(figsize=(10,6))  # Slightly increased figure size
sns.swarmplot(x="species", y="sepal_width", data=iris, size=8, palette="Set2", edgecolor="black", linewidth=1)  # Changed color palette, added edgecolor and line width
plt.title("Enhanced Swarm Plot: Sepal Width by Species", fontsize=16)  # More descriptive title with increased font size
plt.xlabel("Species", fontsize=12)  # Enlarged X-axis label font
plt.ylabel("Sepal Width", fontsize=12)  # Enlarged Y-axis label font
plt.grid(True, linestyle='--', alpha=0.6)  # Added grid for better readability
plt.tight_layout()  # Ensure proper layout
plt.show()
```

##  A swarm plot is a type of categorical scatter plot where individual data points are represented, and they are adjusted to avoid overlap.

## 10. Area Plot

```python
# Data for area plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Enhanced Area Plot with additional customizations
plt.figure(figsize=(10,6))  # Increased figure size
plt.fill_between(x, y1, color="lightcoral", alpha=0.5, label="sin(x)", edgecolor='red', linewidth=1.5)  # Added edge color and increased alpha
plt.fill_between(x, y2, color="lightseagreen", alpha=0.5, label="cos(x)", edgecolor='green', linewidth=1.5)  # Added edge color and increased alpha
plt.title('Enhanced Area Plot: Sine and Cosine Functions', fontsize=16)  # More descriptive title and increased font size
plt.xlabel('X Axis (Radians)', fontsize=12)  # Updated X-axis label
plt.ylabel('Y Axis (Amplitude)', fontsize=12)  # Updated Y-axis label
plt.legend(loc="upper right", fontsize=12)  # Adjusted legend position and font size
plt.grid(True, linestyle='--', alpha=0.6)  # Added grid lines for better readability
plt.tight_layout()  # Ensured proper layout
plt.show()
```

<p><strong>Explanation:</strong> An area plot is similar to a line plot but with the area between the line and the axis filled in. It helps to visualize the cumulative effect of data over time.</p>
<p><strong>Usage:</strong> Best used when tracking multiple quantities over time and you want to visualize the cumulative effect, such as sales growth, revenue, or any time-series data.</p>

## 11. Hexbin Plot

<p><strong>Explanation:</strong> A hexbin plot displays the relationship between two numerical variables when the data is too dense to plot with individual points. It divides the plot into hexagonal bins and counts the number of points within each bin.</p>
<p><strong>Usage:</strong> Best used for visualizing dense data where individual points would overlap, such as in large datasets with many data points. It helps identify areas of high concentration.</p>

```python
# Data for hexbin plot
x = np.random.randn(1000)
y = np.random.randn(1000)

# Enhanced Hexbin Plot with additional customizations
plt.figure(figsize=(10,6))  # Slightly increased figure size
plt.hexbin(x, y, gridsize=50, cmap='Reds', edgecolors='grey', linewidths=0.5)  # Changed color map, added edgecolors and linewidths
plt.colorbar(label='Count in Bin')  # Added more descriptive color bar label
plt.title('Enhanced Hexbin Plot: Distribution of Random Data', fontsize=16)  # Enhanced title with larger font and more description
plt.xlabel('X Axis (Random Values)', fontsize=12)  # Added label to X axis with description
plt.ylabel('Y Axis (Random Values)', fontsize=12)  # Added label to Y axis with description
plt.tight_layout()  # Ensures proper layout and display
plt.show()
```

 <p><strong>Explanation:</strong> A hexbin plot displays the relationship between two numerical variables when the data is too dense to plot with individual points. It divides the plot into hexagonal bins and counts the number of points within each bin.</p>
<p><strong>Usage:</strong> Best used for visualizing dense data where individual points would overlap, such as in large datasets with many data points. It helps identify areas of high concentration.</p>


## 12. Pair Grid (Using Seaborn)

<p><strong>Explanation:</strong> A pair grid is an advanced version of a pair plot. It allows you to create a matrix of plots, showing the relationships between variables in a dataset. You can customize the upper, lower, and diagonal parts of the grid to display different types of plots.</p>
<p><strong>Usage:</strong> Useful when you need more control over how different relationships between variables are visualized. Itâ€™s often used in exploratory data analysis to visualize interactions between features in a dataset.</p>

```python
# Modified Pair Grid with custom settings
g = sns.PairGrid(iris, hue="species", palette="coolwarm", diag_sharey=False)  # Added palette and turned off shared y-axis on diagonals
g.map_upper(sns.scatterplot, edgecolor='black', s=60)  # Added edge color and increased marker size for scatter plot in upper grid
g.map_lower(sns.kdeplot, cmap="Blues_d", fill=True)  # Added color and filled area for KDE plot in lower grid
g.map_diag(sns.histplot, kde=True, color='lightblue')  # Removed kde_kws with 'color' and 'linewidth'
g.add_legend(title="Species", frameon=True)  # Added title to legend and frame
plt.suptitle("Enhanced PairGrid: Iris Dataset", fontsize=16, y=1.02)  # Added a title with a larger font size and adjusted position
plt.tight_layout()  # Ensure proper layout
plt.show()
```

## 13. Facet Grid (Using Seaborn)

<p><strong>Explanation:</strong> A facet grid allows you to create multiple subplots, each visualizing a subset of the data based on a particular categorical variable. This allows you to compare trends across different subsets of the data.</p>
<p><strong>Usage:</strong> Best used for visualizing relationships in data subsets. Itâ€™s especially helpful when you have a categorical variable that you want to use to split the data into multiple panels for comparison (e.g., comparing data across different groups or classes).</p>

```python
# Modified Facet Grid with additional customizations
g = sns.FacetGrid(iris, col="species", col_wrap=3, height=4, aspect=1.2, palette="Set1")  # Adjusted height, aspect ratio, and added palette
g.map(sns.scatterplot, "sepal_length", "sepal_width", edgecolor="black", s=80)  # Added edgecolor and increased marker size for clarity
g.set_axis_labels("Sepal Length (cm)", "Sepal Width (cm)")  # Set custom axis labels
g.set_titles(col_template="{col_name} Species", size=14)  # Customized facet titles with larger font size
g.add_legend()  # Add legend if needed for hue (though not used here)
plt.subplots_adjust(top=0.9)  # Adjust subplot layout for better spacing
g.fig.suptitle("Enhanced FacetGrid: Sepal Length vs Sepal Width", fontsize=16)  # Added a super title with larger font
plt.show()
```
